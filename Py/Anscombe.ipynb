```python
import pandas as pd 
import numpy as np 
import statsmodels.api as sm 
import matplotlib.pyplot as plt 
import seaborn as sns
import timeit
```

```python
anscombe = sns.load_dataset("anscombe", cache=False)
#print(anscombe)
```

```python
# Define the function to be benchmarked
def task():
    set_I_data = anscombe[anscombe['dataset'] == 'I']
    set_I_design_matrix = sm.add_constant(set_I_data['x'])
    set_I_model = sm.OLS(set_I_data['y'], set_I_design_matrix)
    set_I_results = set_I_model.fit()

    set_II_data = anscombe[anscombe['dataset'] == 'II']
    set_II_design_matrix = sm.add_constant(set_II_data['x'])
    set_II_model = sm.OLS(set_II_data['y'], set_II_design_matrix)
    set_II_results = set_II_model.fit()

    set_III_data = anscombe[anscombe['dataset'] == 'III']
    set_III_design_matrix = sm.add_constant(set_III_data['x'])
    set_III_model = sm.OLS(set_III_data['y'], set_III_design_matrix)
    set_III_results = set_III_model.fit()

    set_IV_data = anscombe[anscombe['dataset'] == 'IV']
    set_IV_design_matrix = sm.add_constant(set_IV_data['x'])
    set_IV_model = sm.OLS(set_IV_data['y'], set_IV_design_matrix)
    set_IV_results = set_IV_model.fit()
    
    # Return summary tables for each set
    return set_I_results.summary(), set_II_results.summary(), set_III_results.summary(), set_IV_results.summary()

# Benchmark the task and directly get the average time
average_time = timeit.timeit('task()', setup='from __main__ import task', number=1000) / 1000

# Print the average time
print(f'Average time: {average_time:.6f} seconds')
```
